# HW 2

## ОПИСАНИЕ РАБОТЫ

были произведены замеры производительности для двух политик замещения в L2 кэше: LRU и LFU. результаты измерений будут описаны и проанализированы ниже.

## ПРЕДПОЛОЖЕНИЕ

обе политики опираются на чисто практические наблюдения - LRU на то, что если мы обратились к элементу, то мы скоро обратимся к нему еще, а LFU - на то, что чем чаще мы к элементу обращаемся, тем больше вероятность, что мы к нему обратимся еще раз. предположение - данные политики не должны показывать большую разницу в performance на бенчмарках.

## РЕЗУЛЬТАТЫ

из полученных графиков видно, что смена политики в L2 кэше действительно не привела к существенным изменениям в производительности. рассмотрим результаты более подробно.

### IPC

рассмотрим IPC, как наиболее наглядный показатель производительности.

![IPC](out-bck/ipc.png "IPC")

из графика видно, что в среднем смена политики не привела к существенным изменениям IPC.

### L1 КЭШ (DATA)

рассмотрим, как поменялось поведение L2 кэша при смене политики.

![L1](out-bck/l1d_statistics.png "L1")

на этом графике заметно, что заметные изменения в поведении L2 кэша произходили там, где изменялось поведение L1, что логично. смена политики L2 могла повлиять на L1 следующим образом: с новой политикой часто используемые объекты дольше задерживаются в L2, что и приводит к изменению HIT% в обоих кэшах.

### L2 КЭШ

рассмотрим, как поменялось поведение L2 кэша при смене политики.

![L2](out-bck/l2_statistics.png "L2")

на этом графике уже заметны последствия смены политики. разница между HIT% у LRU и LFU достигает на некоторых бенчмарках 3%, однако в среднем эти политики все еще ведут себя одинаково, что подтверждает изначальное предположение.

### LLC (LAST LEVEL CACHE) КЭШ

теперь рассмотрим поведение последнего уровня кэша - LLC:

![LLC](out-bck/llc_statistics.png "LLC")

из графика видно, что количество доступов и HIT% заметно изменились там, где изменилось поведение L2, причем изменения диаметрально противоположные - где количество доступов в L2 и HIT% упали, там они выросли в LLC, что является ожидаемым поведением. как и раньше, в среднем все показатели остались теми же самыми относительно LRU, однако в бенчмарках, где видны заметные изменения мы видим увеличение доступов в LLC, что говорит о том, что на данных бенчмарках смена политики оказалась пагубной, т.к. время доступов увеличилось (LLC самый большой по памяти кэш). однако в бенчмарке 644 смена политики привела к увеличению HIT% в L2 кэше, что повлекло за собой заметное уменьшение доступов в LLC. для этого бенчмарка смена политики оказалась полезной. стоит заметить, что для этого бенчмарка был зафиксирован самый большой прирост IPC (да. он небольшой, но корреляция есть!)
